```python
import pandas as pd
```

```python
#read csv file
game_df = pd.read_csv("vgsales.csv")
game_df.head()
```

```python
game_df.dtypes
```

```python
game_df.describe()
```

```python
top_df = game_df[["Rank", "Name","Year", "NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales"]] \
    .groupby("Name") \
    .sum() \
    .sort_values(by = "Global_Sales", ascending=False)
top_df.head()
```

```python
end_df = game_df[["Rank", "Name","Year", "NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales"]] \
    .groupby("Name") \
    .sum() \
    .sort_values(by = "Global_Sales", ascending=True)
end_df.head()
```

```python
cleaned_df = game_df.dropna(subset=["NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales"])
cleaned_df.tail()
```
